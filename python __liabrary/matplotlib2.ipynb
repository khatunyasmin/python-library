```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [10, 15, 7, 12]}

df = pd.DataFrame(data)
```

```python
df.plot(x='Category', y='Values', kind='bar', color='g')

# You can also set the title and labels for the chart:
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the plot
plt.show()
```

```python
# Specify the x and y values for the bar chart
x = df['Category']
y = df['Values']

# Create the bar chart
plt.bar(x, y, color='g')

# Set the title and labels for the chart
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the plot
plt.show()
```
