# KDE plot
## KDE plot is a Kernel Density Estimate that is used for visualizing the Probability Density of the continuous or non-parametric data variables

```python

# Assuming you have loaded the "iris" dataset
iris = sns.load_dataset("iris")

# Plotting the KDE Plot
sns.kdeplot(iris.loc[iris['species'] == 'setosa', 'sepal_length'], color='b', fill=True, label='setosa')
sns.kdeplot(iris.loc[iris['species'] == 'virginica', 'sepal_length'], color='r', fill=True, label='virginica')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python

# Setting up the samples
iris_setosa = iris.query("species=='setosa'")
iris_virginica = iris.query("species=='virginica'")

# Plotting the KDE Plot for sepal_length and sepal_width
sns.kdeplot(data=iris_setosa, x='sepal_length', y='sepal_width', cmap="Reds", fill=True, label='Iris Setosa', common_norm=False, thresh=0.05)
sns.kdeplot(data=iris_virginica, x='sepal_length', y='sepal_width', cmap="Blues", fill=True, label='Iris Virginica', common_norm=False, thresh=0.05)

# Set labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('KDE Plot for Sepal Length and Sepal Width')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

# Cluster map
## Cluster map method plots a matrix dataset as a hierarchically-clustered heatmap. It uses hierarchical clusters to order data by similarity

```python

flights=sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
sns.clustermap(flights,linewidths=.5,cmap="coolwarm")
```

```python
flights=sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
sns.clustermap(flights,linewidths=.5,cmap="coolwarm",col_cluster=False)
```

# Facetgrid
## Facet grid forms a matrix of panels defined by row and column by dividing the variables. Due to panels, a single plot looks like multiple plots. It is very helpful to analyze all combinations in two discrete variables.

```python
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.scatterplot, "total_bill", "tip")
```
