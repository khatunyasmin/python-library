# Concatenate DataFrames

```python
india_weather = pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city":["newyork","chicago","orlando"],
    "temperature":[21,15,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
concate = pd.concat([india_weather,us_weather]) #concate two dataframes
print(concate)
```

```python
concate = pd.concat([india_weather,us_weather],ignore_index=True) # continuous index or irdered index
print(concate)
```

```python
side = pd.concat([india_weather,us_weather],axis=1) #print both data side by side
print(side)
```

# Merge DataFrames

```python
temp_df = pd.DataFrame({
    "city":["mumbai","delhi","bangalore","hyderabad"],
    "temperature":[32,45,30,40]
})
temp_df
```

```python
humidity_df = pd.DataFrame({
    "city":["delhi","mumbai","bangalore"],
    "humidity":[68,68,75]
})
humidity_df
```

```python
merge = pd.merge(temp_df,humidity_df,on='city')
merge
```

```python
merge = pd.merge(temp_df,humidity_df,on='city',how='outer')
merge
```
