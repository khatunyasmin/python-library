```python
division=["a","b","c","d","e"]
marks=[92,80,70,65,50]
plt.bar(division,marks,color="green")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
division=["a","b","c","d","e"]
marks=[92,84,90,75,60]
plt.barh(division,marks,color="red")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

division = ["a", "b", "c", "d", "e"]
marks = [92, 84, 80, 75, 60]

# Create a pie chart
plt.pie(marks, labels=division, autopct='%1.1f%%')#is used to display the percentage on each slice.
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.legend(title="List of divisons")
plt.title("Marks Distribution")
plt.show()
```
