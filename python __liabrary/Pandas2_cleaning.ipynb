# Empty Cells

```python
# return new dataframe withput the row of empty values
new_df = df.dropna()
print(new_df.to_string())
```

```python
#it will not return a new dataframe, it will remove null valued rows
df.dropna(inplace=True)
print(df.to_string())
```

```python
#replace empty cells with a value
x=df['culmen_length_mm'].mean()
df['culmen_length_mm'].fillna(x,inplace=True)
print(df.to_string())
```

```python
#replace empty cells with a value
x=df['culmen_length_mm'].mode()
print(x)
df['culmen_length_mm'].fillna(x,inplace=True)
print(df.to_string())
```

```python
#replace empty cells with a value
x=df['culmen_length_mm'].median()
df['culmen_length_mm'].fillna(x,inplace=True)
print(df.to_string())
```
