## Vertical Barplot

```python
sns.set_context('paper')
titanic=sns.load_dataset('titanic')
titanic
```

```python
sns.barplot(x='embark_town',y='age',data=titanic,palette='PuRd',errorbar=None)
plt.legend()
plt.show()
```

## Horizontal barplot

```python
sns.barplot(y='embark_town',x='age',data=titanic,palette='PuRd',orient='h',errorbar=None)
plt.legend()
plt.show()
```

## countplot

```python
sns.set_context('paper')
sns.countplot(x='class',hue='who',data=titanic,palette='magma')
plt.title('survi')
plt.show()
```
