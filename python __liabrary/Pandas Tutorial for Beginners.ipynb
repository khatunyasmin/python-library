# merging and concate

```python
y=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
y1=pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.merge(y,y1,on="A")
```

```python
y=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
y1=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(y,y1,on="A")
```

```python
y=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
y1=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(y,y1,how="inner")
```

```python
y=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
y1=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(y,y1,how="left")
```

```python
y=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
y1=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(y,y1,how="right")
```

```python
y=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
y1=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(y,y1,how="outer")
```

```python
y=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
y1=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(y,y1,how="outer",indicator=True)
```

```python
y=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
y1=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.merge(y,y1,left_index=True,right_index=True)
```

```python
#change the column name
pd.merge(y,y1,left_index=True,right_index=True,suffixes=("name","id"))
```
