# 3. How to Plot a Line Chart in Python

```python
x=["sila","rahul","rohit","yasu"]
y=[85,70,90,82]
c=["r","y","g","b"]
plt.bar(x,y,color=c)
plt.show()
```

# add in label

```python
x=["sila","rahul","rohit","yasu"]
y=[85,70,90,82]
c=["r","y","g","b"]
plt.xlabel("Name")
plt.ylabel("Number")
plt.bar(x,y,color=c)
plt.show()
```

# add in title

```python
x=["sila","rahul","rohit","yasu"]
y=[85,70,90,82]
c=["r","y","g","b"]
plt.xlabel("Name")
plt.ylabel("Number")
plt.title("practice")
plt.bar(x,y,color=c)
plt.show()
```

# change in font size

```python
x=["sila","rahul","rohit","yasu"]
y=[85,70,90,82]
c=["r","y","g","b"]
plt.xlabel("Name",fontsize=25)
plt.ylabel("Number",fontsize=25)
plt.title("practice",fontsize=25)
plt.bar(x,y,color=c)
plt.show()
```

#  change in width

```python
x=["sila","rahul","rohit","yasu"]
y=[85,70,90,82]
c=["r","y","g","b"]
plt.xlabel("Name",fontsize=25)
plt.ylabel("Number",fontsize=25)
plt.title("practice",fontsize=25)
plt.bar(x,y,color=c,width=0.2)
plt.show()
```

# use align parameter

```python
x=["sila","rahul","rohit","yasu"]
y=[85,70,90,82]
c=["r","y","g","b"]
plt.xlabel("Name",fontsize=25)
plt.ylabel("Number",fontsize=25)
plt.title("practice",fontsize=25)
plt.bar(x,y,color=c,width=0.2,align="edge")
plt.show()
```

```python
plt.bar(x,y,color=c,width=0.2,align="center")
plt.show()
```

# change in edge color & linewidth

```python
x=["sila","rahul","rohit","yasu"]
y=[85,70,90,82]
c=["r","y","g","b"]
plt.xlabel("Name",fontsize=25)
plt.ylabel("Number",fontsize=25)
plt.title("practice",fontsize=25)
plt.bar(x,y,color=c,width=0.2,edgecolor="r",linewidth=5)
plt.show()
```

```python
# add in linestyle
```

```python
plt.bar(x,y,color=c,width=0.2,edgecolor="r",linewidth=5,linestyle=":")
plt.show()
```

# change in color light then use in it alpha  its value stari in 0.1 to 1

```python
plt.bar(x,y,color=c,width=0.2,edgecolor="r",linewidth=5,linestyle=":",alpha=0.4)
plt.show()
```

# use in label and legend

```python
plt.bar(x,y,color="r",width=0.2,edgecolor="r",label="practice")
plt.legend()
plt.show()
```

# create a mupltiple bargraph

```python
z=[20,30,40,50]
plt.bar(x,y,color="r",width=0.2,edgecolor="r",label="practice")
plt.bar(x,z,color="y",width=0.2,edgecolor="y",label="practice1")
plt.legend()
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[3,2,1,6,5,4,8]
plt.scatter(day,no)
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[3,2,1,6,5,4,8]
plt.scatter(day,no,color="r")
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[3,2,1,6,5,4,8]
colors=["r","g","b","y","r","b","g"]
plt.scatter(day,no,c=colors)
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[3,2,1,6,5,4,8]
colors=["r","g","b","y","r","b","g"]
sizes=[400,200,300,150,450,200,250]
plt.scatter(day,no,c=colors,s=sizes)
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.show()
```

# use in alpha

```python
day=[1,2,3,4,5,6,7]
no=[3,2,1,6,5,4,8]
colors=["r","g","b","y","r","b","g"]
sizes=[400,200,300,150,450,200,250]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.4)
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.show()
```

```python
# marker
```

```python
day=[1,2,3,4,5,6,7]
no=[3,2,1,6,5,4,8]
colors=["r","g","b","y","r","b","g"]
sizes=[400,200,300,150,450,200,250]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.4,marker="*")
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.show()
```

```python
# edgcolor
```

```python
day=[1,2,3,4,5,6,7]
no=[3,2,1,6,5,4,8]
colors=["r","g","b","y","r","b","g"]
sizes=[400,200,300,150,450,200,250]
plt.scatter(day,no,c=colors,s=sizes,marker="*",edgecolor="g")
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.show()
```

```python
# linewidth
```

```python
day=[1,2,3,4,5,6,7]
no=[3,2,1,6,5,4,8]
colors=["r","g","b","y","r","b","g"]
sizes=[400,200,300,150,450,200,250]
plt.scatter(day,no,c=colors,s=sizes,marker="*",edgecolor="g",linewidth=3)
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.show()
```
